```python
#MNSIT
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

```python
data=pd.read_csv('mnist_train.csv',header=None)
data_test=pd.read_csv('mnist_test.csv',header=None)
data.head()
data_test.head()
```

```python
sample=data.sample(5000,random_state=1)
y=sample[0]
x=sample.iloc[:,1:] 
#y=data[0]
#x=data.iloc[:, 1:] 
pt.figure(figsize=(7,7))
idx = 1200
grid_data=x.iloc[idx].values.reshape(28,28)
pt.imshow(grid_data,interpolation="nearest",cmap="gray")
pt.show()
print(y.iloc[idx])
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
#Logictic regression
prediction=model.predict(x_test)
print(classification_report(y_test,prediction))
print(accuracy_score(y_test,prediction))
```

```python
from sklearn import svm
from sklearn.svm import SVC
model=SVC(kernel="rbf",random_state=42)
model.fit(x_train,y_train)
prediction=model.predict(x_test)
print(classification_report(y_test,prediction))
print(accuracy_score(y_test,prediction))
```
